### Tri par selection
En pratique, lorsque nous voulons trier, nous utiliserons généralement l'une des deux méthodes que nous avons apprises dans le cours précédent. Cependant, étant donné que le tri est une tâche fondamentale, il est bon de savoir comment le tri fonctionne dans les coulisses.

Il existe de nombreux algorithmes de tri ; les enseigner tous prendrait probablement pas mal de leçons. Dans cette leçon, nous allons examiner un algorithme nommé tri par sélection. Cet algorithme n'est pas le plus efficace, mais l'analyse de sa complexité nous apprendra des techniques intéressantes que nous pourrons ensuite appliquer à d'autres algorithmes.

La plupart des algorithmes de tri sont basés sur l'idée d'échanger des éléments. Étant donné une liste et deux indices de liste, **i** et **j**, permuter les positions **i** et **j** consiste à remplacer la valeur à l'indice **i** par la valeur à l'indice **j** et inversement.

#### Strategie de tri par selection

Nous sommes maintenant prêts à reprendre le tri. Si vous pensez à une liste triée, le premier élément sera la valeur minimale de la liste et le deuxième élément sera le minimum des éléments restants, et ainsi de suite. Ainsi, une stratégie pour trier consiste à :

1. Trouvez l'index de l'élément minimum et échangez-le avec l'élément à l'index 0.

Capture d’écran 2022-02-11 à 09.55.32.png

2. Trouver l'index du minimum pour le reste de la liste et echanger le avec l'element qui est à l'indice 1
Capture d’écran 2022-02-11 à 09.57.38.png

3. Trouver le minimum dans le reste de la liste et l'echanger avec l'element d'indice 2 et ainsi de suite...
Capture d’écran 2022-02-11 à 09.59.28.png

Comme vous pouvez le voir, à chaque étape, nous recherchons l'indice de la valeur minimale. D'abord, sur les indices [0, 5], puis [1, 5], puis [2, 5], et ainsi de suite. La seule chose qui change à chaque itération est le début de la plage.

Commençons par écrire une fonction qui, étant donné un tableau et un index de départ, trouve l'index de la valeur minimale dans cette plage.

````python
def select_index_min(liste,index_min):
  mini = liste[index_min]
  i = index_min
  idx_mini = None
  N = len(liste)
  while i < N :
   
    if idx_mini == None or liste[i] < mini:
      mini = liste[i]
      idx_mini = i
    i = i + 1
  return idx_mini
````

````python
liste = [5,4,6,1,3,2]
idx = select_index_min(liste,0)
swap(liste,idx,0)
print(liste)
idx = select_index_min(liste,1)
swap(liste,idx,1)
print(liste)
idx = select_index_min(liste,2)
swap(liste,idx,2)
print(liste)
idx = select_index_min(liste,3)
swap(liste,idx,3)
print(liste)
idx = select_index_min(liste,4)
swap(liste,idx,4)
print(liste)
idx = select_index_min(liste,5)
swap(liste,idx,5)
print(liste)
````

### Exercice 2
Ecrire la fonction de tri selection_sort() qui prend en paramétre une liste et retourne une liste triée

````python
def selection_sort(values):
    idx_max = len(values)
    
    for i in range(idx_max):
        
        idx_min = select_index_min(values,i)
        swap(values,idx_min,i)

liste = [5,4,6,1,3,2]
selection_sort(liste)
print(liste)
````

L'algorithme que nous venons d'implémenter s'appelle le tri par sélection. La raison en est qu'il conserve une plage initiale d'index qui sont triés et, à chaque itération, il sélectionne l'élément qui vient ensuite dans l'ordre et le place à la bonne position.

#### Exercice 4:
Calculer la complexite de cet algorithme

````python
# Complexite select_index_min
def select_index_min(liste,index_min):
  mini = liste[index_min]    #1
  i = index_min   #1
  idx_mini = None  #1
  N = len(liste) #1
  while i < N :  #N
   
    if idx_mini == None or liste[i] < mini:  #2N
      mini = liste[i]  #N
      idx_mini = i #N
    i = i + 1 #N
  return idx_mini #1
````

````python
def selection_sort(values):
    idx_max = len(values)
    
    for i in range(idx_max): #N
        
        idx_min = select_index_min(values,i) #depend de i: N si i = 0, N-1 si i = 1,...,1 si i = N-1
        swap(values,idx_min,i) #N
````

La complexite de selection_sort est equivalent à N + (N-1)+(N-2),+....+1.

#### Exercice 8
Comparer les temps d'exécution de la fonction **selection_sort()** que vous avez implémenté avec la fonction **sorted()** de python.
Vous pouvez générer une liste aléatoire de nombre et utiliser la fonction **time()**

````python
import matplotlib.pyplot as plt
import random

def plot_times(times_python, times_selection):
    plt.plot(times_python, label='time sorted()')
    plt.plot(times_selection, label='time selection_sort()')
    plt.ylabel('runtime')
    plt.xlabel('N')
    plt.legend()
    plt.show()


# Write your code below
import time
times_python = []
times_selection = []

for N in range(1,501):
    rnd_values = [random.randint(1, 10000) for _ in range(N)]
    start = time.time()
    sorted(rnd_values)
    end = time.time()
    runtime = end - start
    times_python.append(runtime)
    
    start = time.time()    

    
    selection_sort(rnd_values)
    end = time.time()
    runtime = end - start
    times_selection.append(runtime)
    
plot_times(times_python, times_selection)
````

Comme vous l'avez vu dans l'expérience précédente, l'algorithme de tri en Python est beaucoup plus efficace que le tri selection.

En fait, il est de complexité $O(Nlog(N))$, alors que notre algorithme est de complexité $O(N^2)$. Rappelons que $log(N)$ croît beaucoup plus lentement que $N$ donc, un algorithme $O(N log(N))$ sera beaucoup plus rapide qu'un algorithme $O(N^2)$.

La complexité temporelle $O(N log(N))$ se situe entre $O(N)$ et $O(N^2)$ mais est beaucoup plus proche de $O(N)$ que de $O(N^2)$
