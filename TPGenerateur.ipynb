## Exemples d'utilisation

```python
def generateur_nombre():
  num = 0
  while True:
    yield num
    num += 1
```

### Methodes avancées

**throw** permet de lancer une exception à partir du générateur

```python
def exceptiongenerator(text):
  for i in text:
    ret = yield i
    print(ret)
    
```

```python
gen_car = exceptiongenerator('biaise')
for i in gen_car:
  if i == 'a':
      gen_car.throw(ValueError("caractere 'a' non autorise"))
  else:
    print(i)
```

```python
gen_car1 = exceptiongenerator('biaise')
for i in gen_car1:
  if i == 'a':
      gen_car1.close()
  else:
    print(i)
```

```python
import os
def all_lines(path):
    for filename in os.listdir(path):
        full_filename = os.path.join(path,filename)
        try:
            for line in open(full_filename):
                yield line
        except OSError:
            pass
```

**Assembler les codes et creer un pipeline**

```python

lines = (line for line in open("/content/drive/MyDrive/Colab Notebooks/Cours/FormationPython/test_generateur.csv",encoding= 'unicode_escape'))
list_line = (s.rstrip().split(",") for s in lines)
cols = next(list_line)

company_dicts = (dict(zip(cols, data)) for data in list_line)

funding = (
    int(company_dict["raisedAmt"])
    for company_dict in company_dicts
    if company_dict["round"] == "a"
)
total_series_a = sum(list(funding))


print(f"Total series A fundraising: ${total_series_a}")
```

Exercice: le montant moyen levé par les compagnies dans un round a


*   Les générateurs s'épuisent après avoir été complètement itérés.
*   Vous aurez besoin de la fonction sum



**Solution**

```python
import itertools
lines = (line for line in open("/content/drive/MyDrive/Colab Notebooks/Cours/FormationPython/test_generateur.csv",encoding= 'unicode_escape'))
list_line = (s.rstrip().split(",") for s in lines)
cols = next(list_line)

company_dicts1, company_dicts2 = itertools.tee((dict(zip(cols, data)) for data in list_line))

funding = (
    int(company_dict["raisedAmt"])
    for company_dict in company_dicts1
    if company_dict["round"] == "a"
)
total_series_a = sum(list(funding))

count_company_A = len(set(list(company["company"] for company in company_dicts2 if company["round"] == "a")))
print(count_company_A)

print(f"Total series A fundraising: ${total_series_a/count_company_A}")
```
