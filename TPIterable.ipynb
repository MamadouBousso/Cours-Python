```python
class IterListe:

  def __init__(self,liste):
    self.liste = liste
    

  def __enter__(self):
    self.iterateur = iter(self.liste)
    return self

  def __exit__(self,exct,exct_val,exct_tb):
    pass
    
  
  def parcourir(self):
    while True:
      try:
        number = next(self.iterateur)
      except StopIteration:
        break
      print(number)
```

```python
with IterListe([12,14,15]) as it:
  it.parcourir()
```

```python
class MyRange:
  def __init__(self,size):
    self.size = size
  def __iter__(self):
    return MyRangeIterator(self)

class MyRangeIterator:
  def __init__(self,myrange):
    self.current = 0
    self.max = myrange.size

  def __next__(self):
    
    if self.current >= self.max:
      raise StopIteration
    ret = self.current
    self.current = self.current + 1
    return ret

  def __iter__(self):
    return self
```

```python
myrange = MyRange(8)
it = iter(myrange)
next(it)
```

```python
next(it)
```

```python
for i in MyRange(10):
  print(i)
```

```python
class Repeater:
  def __init__(self,chaine,limit):
    self.chaine = chaine
    self.limit = limit
  def __iter__(self):
    return RepeaterIterator(self)
```

```python
class RepeaterIterator:
  def __init__(self,repeater):
    self.repeater = repeater
    self.limit = repeater.limit
    self.current = 0
  def __next__(self):
    if self.current >= self.limit:
      raise StopIteration
    self.current += 1
    return self.repeater.chaine

```

```python
repeater = Repeater('Hello',10)
for item in repeater:
    print(item)
```

### Pourquoi ne pas utiliser une seule classe au lieu de deux ?

```python
# RepaterNew devient iterable et iterator
class RepeaterNew:
  
  def __enter__(self):
    return self
  
  def __exit__(self,exc1,exc2,exc3):
    if isinstance(exc2, StopIteration):
      print("Fin de l'iteration")
      return True
    return False
  
  def __init__(self,chaine,limit):
    self.chaine = chaine
    self.limit = limit
    self.current = 0
  
  def __iter__(self):
    return self

  def __next__(self):
    if self.current >= self.limit:
      raise StopIteration
    self.current += 1
    return self.chaine
```

```python
repeater = RepeaterNew('Hello',10)
for item in repeater:
    print(item)

```

```python
with RepeaterNew('Hello',10) as rep:
  while True:
    print(next(rep))
```
