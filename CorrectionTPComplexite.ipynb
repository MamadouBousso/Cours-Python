```python
def find_maximum(values):
    maximum = None                              # 1
    for value in values:                        # N
        if maximum == None or value > maximum:  # N
            maximum = value                      # N
    return maximum                               # 1
```

```python
def find_minimum(values):
    minimum = None                                #1
    for value in values:                          # N
        if minimum == None or value < minimum:    # N
            minimum = value                       # N
    return minimum                                # 1
```

```python

def min_max2(values):
    minimum = find_minimum(values)               # N
    maximum = find_maximum(values)               # N
    return minimum, maximum                       # 1
```

```python
import matplotlib.pyplot as plt

def plot_costs(times,double_times,triple_times):
    plt.plot(times, label='append cost')
    plt.plot(double_times, label='2N')
    plt.plot(triple_times, label='3N')
    plt.ylabel('cost')
    plt.xlabel('N')
    plt.show()
    
def append_cost(array_length, list_length):
    if array_length == list_length:
        return array_length
    return 1

def append_N_list_cost(N):
    array_length = 1 # initially the array will have length 1
    list_length = 0 # initially the list has 0 elements
    total_cost = 0 # this variable will keep track of the total cost
    for i in range(N):
        cost = append_cost(array_length, list_length)
        total_cost += cost
        # update the array and list lengths
        if array_length == list_length:
            array_length *= 2
        list_length += 1
    return total_cost

# add you code below
costs = []
double_costs = []
triple_costs = []

for i in range(5000):
    cost = append_N_list_cost(i)
    costs.append(cost)
    double_costs.append(2*i)
    triple_costs.append(3*i)
    
plot_costs(costs,double_costs,triple_costs)
#plot_costs(double_costs)
#plot_costs(triple_costs)
```

```python
import random


def randN(N):
	min = pow(10, N-1)
	max = pow(10, N) - 1
	return random.randint(min, max)
paires = [(randN(i),randN(i)) for i in range(1,10000)]
```

```python
import matplotlib.pyplot as plt
import time
def plot_times(times):
    plt.plot(times)
    plt.ylabel('multiplication runtime')
    plt.xlabel('number of digits')
    plt.yscale('log')
    plt.yticks([])
    plt.show()

times = []
for couple in paires:
    start = time.time()
    couple[0]*couple[1]
    end  = time.time()
    times.append(end-start)
plot_times(times)
```
